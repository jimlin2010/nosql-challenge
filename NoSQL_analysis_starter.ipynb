```python
# Import dependencies

from pymongo import MongoClient
from pprint import pprint

from decimal import Decimal
import pymongo

import pandas as pd
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}
hygiene_score = establishments.find(query)

# Use count_documents to display the number of documents in the result

establishments_number = establishments.count_documents(query)
print(establishments_number)


# Display the first document in the results using pprint

pprint(hygiene_score[0])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_score_df = pd.DataFrame(list(hygiene_score))

# Display the number of rows in the DataFrame
print(f'Number of rows: {len(hygiene_score_df)}')

# Display the first 10 rows of the DataFrame
hygiene_score_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.

#Cannot find any local authority name with "London", so use AddressLine3 instead
query = {'AddressLine3': 'London', 'RatingValue': {'$gte': '4', '$lte': '6'}}
london_rating = establishments.find(query)


# Use count_documents to display the number of documents in the result

london_rating_number = establishments.count_documents(query)
print(london_rating_number)

# Display the first document in the results using pprint

pprint(london_rating[0])
```

```python
# Convert the result to a Pandas DataFrame
london_rating_df = pd.DataFrame(list(london_rating))

# Display the number of rows in the DataFrame
print(f'Number of rows: {len(london_rating_df)}')

# Display the first 10 rows of the DataFrame
london_rating_df.head(10)

```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.4901420000000
longitude = 0.0838400000000000

query = {'geocode.latitude': {'$gte': latitude - degree_search, '$lte': latitude + degree_search},
         'geocode.longitude': {'$gte': longitude - degree_search, '$lte': longitude + degree_search},
         'RatingValue': '5'}

sort = [('scores.Hygiene', -1)]

penang_ds = establishments.find(query).sort(sort)


# Print the results
for results in penang_ds:
    pprint(results)

```

```python
penang_ds = establishments.find(query).sort(sort)

# Convert result to Pandas DataFrame

penang_ds_df = pd.DataFrame(list(penang_ds))

# Display the first 10 rows of the DataFrame
penang_ds_df.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

pipeline = [{"$match": {"scores.Hygiene": 0}},
            {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
            {"$sort": {"count": -1}}]

hygiene_score_zero = establishments.aggregate(pipeline)
count_zero = establishments.estimated_document_count()

# Print the number of documents in the result
print("Number of documents in result: ", count_zero)


# Print the first 10 results
for i in hygiene_score_zero:
    pprint(i)


```

```python
# Convert the result to a Pandas DataFrame

hygiene_score_zero = establishments.aggregate(pipeline)
hygiene_score_zero_df = pd.DataFrame(list(hygiene_score_zero))

# Display the number of rows in the DataFrame

print(f'Number of rows: {len(hygiene_score_zero_df)}')

# Display the first 10 rows of the DataFrame

hygiene_score_zero_df.head(10)
```
